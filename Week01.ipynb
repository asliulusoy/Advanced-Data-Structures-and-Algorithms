## Week 01 Exercise

* tinyG.txt from https://algs4.cs.princeton.edu/41graph/tinyG.txt
* networkx & matplotlib for:
    * graph visualization
* undirected_graphs for:
    * degree of vertex
    * maximum degree of vertex
    * number of self-loops in a graph

```python
import networkx as nx
import matplotlib.pyplot as plt
from undirected_graphs import Graph
```

```python
g = Graph.load("tinyG.txt")
print(g)
print(f"Degree of vertex 0: {g.degree(0)}")
print(f"Maximum degree of G: {g.max_degree()}")
print(f"Number of self loops in G: {g.self_loops()}")
```

```python
g_nx = g.to_networkx()
nx.draw_networkx(
    g_nx,
    nx.spring_layout(g_nx, 1, seed=10),
    node_size=300,
    with_labels=True,
    alpha=0.9,
    width=0.5,
    )
plt.show()
```
